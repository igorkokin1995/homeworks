###Обязательная часть
Вам необходимо написать функцию, которая будет основана на поиске по сайту habr.com. Функция в качестве параметра должна принимать список запросов для поиска (например, ['python', 'анализ данных']) и на основе материалов, попавших в результаты поиска по каждому запросу, возвращать датафрейм вида:

###<дата> - <заголовок> - <ссылка на материал>
В рамках задания предполагается работа только с одной (первой) страницей результатов поисковой выдачи для каждого запроса. Материалы в датафрейме не должны дублироваться, если они попадали в результаты поиска для нескольких запросов из списка.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
```

```python
def habr(query, page):
    """
    Функция парсинга по сайту habr.com

    query - поисковые запросы

    page - переход на следующую страницу поиска. Формат 'page2', 'page3'... 
    Для первой указать ''
    """
    habr_serch = pd.DataFrame()
    for q in query:
        URL = f'https://habr.com/ru/search/{page}'
        params = {
            'q': q
        }
        req = requests.get(URL, params=params)
        time.sleep(0.3)
        soup = BeautifulSoup(req.text)
        articles = soup.find_all('article', class_='tm-articles-list__item')

        for article in articles:
          if article.find('h2').find('a'):
            link = 'https://habr.com' + article.find('h2').find('a').get('href')
          elif article.find('div').find('a'):
            link = 'https://habr.com' + article.find('div').find('a').get('href')
  
          title = article.find('h2').text
          date = article.find('time').text.strip()
          like = article.find('svg').text

          row = {'date': date, 'title': title, 'link': link, 'like': like}
          habr_serch = pd.concat([habr_serch, pd.DataFrame([row])]) 
            
    return habr_serch.reset_index(drop=True)

res = habr(['python', 'анализ данных'],'')
res
```
