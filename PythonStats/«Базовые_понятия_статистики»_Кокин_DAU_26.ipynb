###Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
df_horse_new.isna().sum() #кол-во NaN значений
```

```python
df_no_NaN = df_horse_new.copy()
```

```python
df_horse_new['surgery?'].value_counts()
```

Т.к значение 1 в столбце 'surgery?' встречается чаще, логично предположить, что в единственном пропуске должно быть также 1

```python
df_no_NaN['surgery?'] = df_no_NaN['surgery?'].fillna(df_no_NaN['surgery?'].mode()[0])
df_no_NaN['surgery?'].value_counts()
```

Значения температуры у лошадей имеют достаточно маленькое стандартное отклонение и не имеют явных выбросов. Можно предположить, что замена пустых значений на моду в данном столбце не приведет к сильному искажению данных

```python
df_no_NaN['rectal temperature'] = df_no_NaN['rectal temperature'].fillna(df_no_NaN['rectal temperature'].mode()[0])
```

Пульс и частота дыхания могут зависить от того, была ли у лошади операция или нет

```python
df_no_NaN['pulse'] = df_no_NaN['pulse'].fillna(df_no_NaN.groupby('surgery?')['pulse'].transform('median'))
df_no_NaN['respiratory rate'] = df_no_NaN['respiratory rate'].fillna(df_no_NaN.groupby('surgery?')['respiratory rate'].transform('median'))
df_no_NaN
```

```python
df_no_NaN['outcome'].value_counts()
```

Большинство днанных имеют значение 1. логично предположить, что и единственное пропущенное значение также 1.

```python
df_no_NaN['outcome'] = df_no_NaN['outcome'].fillna(df_no_NaN['outcome'].mode()[0])
```

```python
df_no_NaN['pain'].value_counts()
```

Наиболее частое значения для столбца "боль" - 3. Однако в описании данных сказано, что это субьективный параметр. Предсказать такой параметр точно практически невозможно. Данные с пропусками в нем лучше исключить.

Показатель температура конечностей достаточно важен при лечении. Его неверное предсказание может сильно исказить данные.
Предположим, что такие данные лучше исключить.


```python
df_no_NaN = df_no_NaN.dropna()
df_no_NaN
```
